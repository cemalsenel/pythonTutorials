```python
text = "{:.2f}, {:.3f}, {:.4f}".format(3.1456343, 5.479435, 7.3654349)
print(text)
```

```python
text = "{:.2s} - {:.3s} - {:.4s}".format("3.1456343", "5.479435", "7.3654349")
print(text)
```

```python
text = "{:>20}".format("jimmy")  # 20 karakterlik alan içinde sağa yasladı
text
```
